```python
X = df.loc[:, ['Water_(g)', 'Protein_(g)', 'Lipid_Tot_(g)', 'Ash_(g)',
       'Carbohydrt_(g)', 'Fiber_TD_(g)', 'Sugar_Tot_(g)']].values
y = df.loc[:, 'Energ_Kcal'].values
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
X_train.shape
```

```python
X_test.shape
```

# Using Linear Regression, Random forest model and DecissionTree

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_prd = lr.predict(X_test)
print("Linear Regrssion model prediction is ",  lr_prd)
```

```python
from sklearn.ensemble import RandomForestRegressor
rm = RandomForestRegressor()
rm.fit(X_train, y_train)
rm_prd = rm.predict(X_test)
print("Random Forest Model Prediction:", rm_prd)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
print("Decision Tree Classifier Prediction:", dt_pred)
```

# Validating and comparing these models

```python
from sklearn.metrics import mean_absolute_error
```

```python
print("Mean absolute error of Linear Regression", mean_absolute_error(y_test, lr_prd))
```

```python
print("Mean absolute error of Random Forest", mean_absolute_error(y_test, rm_prd))
```

```python
print("Mean absolute error of Decission Tree", mean_absolute_error(y_test, lr_prd))
```
